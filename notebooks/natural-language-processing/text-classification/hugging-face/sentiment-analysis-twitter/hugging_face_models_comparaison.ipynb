```python
#import dependencies
import transformers
import pandas as pd
import re
import torch
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import numpy as np
import seaborn as sns
```

### Load your datasets with Tweets

#### Reel sentiments

```python
df_reel = pd.read_csv("/workspace/data/reel_sentiments_16_10_2021_french_tweets.csv",sep = ';')

# display the first row
df_reel.head()
```

```python
# select the column of interest
label_reel = df_reel['reel_label'].values
df_reel['reel_label'].value_counts()
```

#### Results - BARThez sentiments

```python
df_barthez = pd.read_csv("/workspace/results/results_16_10_2021_french_tweets_barthez.csv",sep = ',')

# display the first row
df_barthez.head()
```

```python
# select the column of interest
label_barthez = df_barthez['label'].values
df_barthez['label'].value_counts()
```

#### Results - CamemBERT sentiments

```python
df_camembert = pd.read_csv("/workspace/results/results_16_10_2021_french_tweets_camembert.csv",sep = ',')

# display the first row
df_camembert.head()
```

```python
# select the column of interest
label_camembert = df_camembert['label'].values

# preprocess: change "POSITIVE" to "Positive" and "NEGATIVE" to "Negative"
k=0

for k in range(len(label_camembert)):
    if label_camembert[k]=="POSITIVE":
        label_camembert[k]="Positive"
    else:
        label_camembert[k]="Negative"
    k+=1
    
df_camembert['label'].value_counts()
```

Now that your data is ready to be used, you are able to compare your results with reality!

### Step 3 - Compare results obtained with BARThez and reel sentiments

```python
i=0

# create 2 variables: 1 for the right results, 1 for the wrong results
false_label_barthez = 0
true_label_barthez = 0

for i in range(len(label_reel)):
    if label_barthez[i]==label_reel[i]:
        true_label_barthez+=1
    else:
        false_label_barthez+=1
    i+=1

# display the two variables: you have the number of right and wrong results
print("True sentiments: ",true_label_barthez)
print("False sentiments: ",false_label_barthez)

# calculate the success rate
success_rate = true_label_barthez*100/(true_label_barthez+false_label_barthez)
print("Success rate: ", success_rate, "%")
```

```python
# display the confusion matrix
cf_matrix_barthez = confusion_matrix(label_reel, label_barthez)
df_cm = pd.DataFrame(cf_matrix_barthez, columns=np.unique(label_reel), index = np.unique(label_reel))
df_cm.index.name = 'Reel sentiment'
df_cm.columns.name = 'Predicted sentiment'
sns.heatmap(df_cm, annot=True, cmap='Blues')
```

### Step 4 - Compare results obtained with CamemBERT and reel sentiments

```python
i=0

# create 2 variables: 1 for the right results, 1 for the wrong results
false_label_camembert = 0
true_label_camembert = 0

for i in range(len(label_reel)):
    if label_camembert[i]==label_reel[i]:
        true_label_camembert+=1
    else:
        false_label_camembert+=1
    i+=1

# display the two variables: you have the number of right and wrong results
print("True sentiments: ",true_label_camembert)
print("False sentiments: ",false_label_camembert)

# calculate the success rate
success_rate = true_label_camembert*100/(true_label_camembert+false_label_camembert)
print("Success rate: ", success_rate, "%")
```

```python
# display the confusion matrix
cf_matrix_camembert = confusion_matrix(label_reel, label_camembert)
df_cm = pd.DataFrame(cf_matrix_camembert, columns=np.unique(label_reel), index = np.unique(label_reel))
df_cm.index.name = 'Reel sentiment'
df_cm.columns.name = 'Predicted sentiment'
sns.heatmap(df_cm, annot=True, cmap='Blues')
```

### Step 5 - Conclusion

In the case of our database, the model based on CamemBERT ([pt-tblard-tf-allocine](https://huggingface.co/philschmid/pt-tblard-tf-allocine)) seems to be the best! Indeed, the numbers of **false positives** and **false negatives** are lower than the model based on BARThez ([barthez-sentiment-classification](https://huggingface.co/moussaKam/barthez)).
